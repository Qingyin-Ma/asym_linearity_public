## Evaluating computation speed via plots 

This file plots (under different parameterization):

- The time taken to solve for the optimal consumption function.

- The number of steps for the time iteration algorithm to terminate.

```python
run optimality
```

```python
tol = 1e-5                                            # tolerance level to terminate time iteration
grid_med = 10                                         # median grid point for saving
sizes = [20, 100, 1000]                               # number of grid points for asset/saving
α_space = exp_grids(gmin=0, gmax=1, gmed=0.1, N=31)   # parameter that controls the initial guess
```

### Parallel Computing

The code has been parallelized, so it runs very fast, and the time taken is relatively less sensitive to the grid size of asset/saving. 

#### Case 1. `grid_max = 30`

```python
speed_plots(α_space=α_space,
            grid_sizes=sizes, 
            grid_max=30,
            grid_med=grid_med, tol=tol, figname='gm_30')
```

#### Case 2. `grid_max = 100`

```python
speed_plots(α_space=α_space,
            grid_sizes=sizes, 
            grid_max=100,
            grid_med=grid_med, tol=tol, figname='gm_100')
```

#### Case 3. `grid_max = 1e6`

```python
speed_plots(α_space=α_space,
            grid_sizes=sizes, 
            grid_max=1e6,
            grid_med=grid_med, tol=tol, figname='gm_1e6')
```

#### Case 4. `grid_max = 1e10`

```python
speed_plots(α_space=α_space,
            grid_sizes=sizes, 
            grid_max=1e10,
            grid_med=grid_med, tol=tol, figname='gm_1e10')
```
