```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x_star(x, γ=3, δ=0.04, r=0.03):
    power = -x*(δ-(1-γ)*r)/γ
    return (1 - np.exp(power))**(-γ)
```

```python
x_space = np.linspace(.08, 1, 100)
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(x_space, x_star(x_space))
ax.set_xlim(0, 1)
ax.set_ylim(1e4, 1e8)
#ax.grid(ls='--', color='grey')
ax.set_xlabel('Length of one period ($\Delta$)', fontsize=12)
ax.set_ylabel('Fixed point ($x^*(z)$)', fontsize=12)
plt.yscale('log')
plt.savefig('fig_xz.pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
